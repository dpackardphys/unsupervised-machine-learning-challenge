```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "myopia.csv"
df = pd.read_csv(file)
df.head()
```

```python
print(df.columns)
```

```python
df = df.drop(['MYOPIC'],axis=1)
normalized = normalize(df)
X = pd.DataFrame(normalized)
X.head()
```

```python
pca = PCA(n_components=0.9)
myopia_pca = pca.fit_transform(X)
# pca.explained_variance_ratio_
df_myopia_pca = pd.DataFrame(myopia_pca)
df_myopia_pca.head()
```

The PCA changed the number of components from 14 to 3.

```python
tsne = TSNE(learning_rate='auto',perplexity=44)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
tsne = TSNE(learning_rate='auto',perplexity=2)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
tsne = TSNE(learning_rate='auto',perplexity=5)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
tsne = TSNE(learning_rate='auto',perplexity=30)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
tsne = TSNE(learning_rate='auto',perplexity=50)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
tsne = TSNE(learning_rate='auto',perplexity=100)
tsne_features = tsne.fit_transform(myopia_pca)
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

There are at least two clear clusters of data visible in several of these plots (e.g. large left blob and smaller right blob in bottom plot). We expect higher values of perplexity to better provide meaningful clustering information, see https://distill.pub/2016/misread-tsne/.

```python
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i)
    km.fit(tsne_features)
    inertia.append(km.inertia_)
    
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

The elbow of the plot is at 2 clusters.

### Conclusion
Based on these findings, the visual clustering seen in the t-SNE plots and the elbow in the KMeans inertia plot, we recommend the patient data be clustered into two clusters.
